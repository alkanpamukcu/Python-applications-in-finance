```python
# Two Dimensional Dataset
np.random.seed(2000)
y = np.random.standard_normal((20,2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(7,4))
plt.plot(y, lw=1.5)
plt.plot(y, 'ro', color='green') #graph is along columns i.e. axis=1
```

```python
# plt.legend accepts different locality parameters
plt.figure(figsize=(7, 4))
plt.plot(y[:, 0], lw=1.5, label='1st') #graph 1st column
plt.plot(y[:, 1], lw=1.5, label='2nd') #graph 2nd column
plt.plot(y, 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

```python
fig, ax1 = plt.subplots()
plt.plot(y[:, 0]*100, 'b', lw=1.5, label ='1st')
plt.plot(y[:, 0]*100, 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')
ax2=ax1.twinx() # plot second data using right axis
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

```python
plt.figure(figsize=(7, 5))
plt.subplot(211) # identify where to put graphs (2 rows 1 column)
plt.plot(y[:, 0]*100, lw=1.5, label='1st')
plt.plot(y[:, 0]*100, 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(212) # same here
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
```

```python
# The first step is the definition
of the function to be integrated:
from matplotlib. patches import Polygon
def func(x):
    return 0.5 * np.exp(x) + 1
a,b = 0.5, 1.5 #integral limits
x = np.linspace(0,2)
y = func(x)

fig, ax = plt.subplots(figsize=(7,5))
plt.plot(x, y, 'b', linewidth=2)
plt.ylim(ymin=0)

# Illustrate the integral value, the area under function

Ix = np.linspace(a, b)
Iy = func(Ix)
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)]
poly = Polygon(verts, facecolor ='0.7', edgecolor = '0.5')
ax.add_patch(poly) # we generate the shaded area (patch) by the use of the Polygon
                   # function illustrating the integral area:
# add text
plt.text(0.5 * (a + b), 1, r"$\int_a^b f(x)\mathrm{d}x$",
horizontalalignment='center', fontsize=25)
plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')
```
